```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
audi = pd.read_csv("../Datasets/Audi.csv", error_bad_lines=False, sep=',', low_memory=False)
audi.drop('Unnamed: 0', axis=1, inplace=True)

kia = pd.read_csv("../Datasets/KIA.csv", error_bad_lines=False, sep=',', low_memory=False)
kia.drop('Unnamed: 0', axis=1, inplace=True)

toyota = pd.read_csv("../Datasets/Toyota.csv", error_bad_lines=False, sep=',', low_memory=False)
toyota.drop('Unnamed: 0', axis=1, inplace=True)

jeep = pd.read_csv("../Datasets/Jeep.csv", error_bad_lines=False, sep=',',low_memory=False)
jeep.drop('Unnamed: 0', axis=1, inplace=True)

hummer = pd.read_csv("../Datasets/Hummer.csv", error_bad_lines=False, sep=',',low_memory=False)
hummer.drop('Unnamed: 0', axis=1, inplace=True)

hyundai = pd.read_csv("../Datasets/Hyundai.csv", error_bad_lines=False, sep=',',low_memory=False)
hyundai.drop('Unnamed: 0', axis=1, inplace=True)

porsche = pd.read_csv("../Datasets/Porsche.csv", error_bad_lines=False, sep=',',low_memory=False)
porsche.drop('Unnamed: 0', axis=1, inplace=True)

superbowl = pd.read_csv("../Datasets/Superbowl.csv", error_bad_lines=False, sep=',',low_memory=False)
superbowl.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
mask_a = (audi['date'] > '2020-02-03 05:30') & (audi['date'] <= '2020-02-03 17:30')
audi_i = audi.loc[mask_a]

mask_a = (kia['date'] > '2020-02-03 05:30') & (kia['date'] <= '2020-02-03 17:30')
kia_i = kia.loc[mask_a]

mask_a = (toyota['date'] > '2020-02-03 05:30') & (toyota['date'] <= '2020-02-03 17:30')
toyota_i = toyota.loc[mask_a]

mask_a = (jeep['date'] > '2020-02-03 05:30') & (jeep['date'] <= '2020-02-03 17:30')
jeep_i = jeep.loc[mask_a]

mask_a = (hummer['date'] > '2020-02-03 05:30') & (hummer['date'] <= '2020-02-03 17:30')
hummer_i = hummer.loc[mask_a]

mask_a = (hyundai['date'] > '2020-02-03 05:30') & (hyundai['date'] <= '2020-02-03 17:30')
hyundai_i = hyundai.loc[mask_a]

mask_a = (porsche['date'] > '2020-02-03 05:30') & (porsche['date'] <= '2020-02-03 17:30')
porsche_i = porsche.loc[mask_a]

mask_a = (superbowl['date'] > '2020-02-03 00:30') & (superbowl['date'] <= '2020-02-03 17:30')
superbowl_i = superbowl.loc[mask_a]
```

```python
audi_c=pd.DataFrame(audi_i.groupby('date').count()['username'])
kia_c=pd.DataFrame(kia_i.groupby('date').count()['username'])
toyota_c=pd.DataFrame(toyota_i.groupby('date').count()['username'])
jeep_c=pd.DataFrame(jeep_i.groupby('date').count()['username'])
hummer_c=pd.DataFrame(hummer_i.groupby('date').count()['username'])
hyundai_c=pd.DataFrame(hyundai_i.groupby('date').count()['username'])
porsche_c=pd.DataFrame(porsche_i.groupby('date').count()['username'])
superbowl_c=pd.DataFrame(superbowl_i.groupby('date').count()['username'])
audi_c
```

```python
together = pd.concat([kia_c, audi_c, toyota_c, jeep_c, hummer_c, hyundai_c, porsche_c], axis=1).fillna(0)
together.columns = ['Kia','Audi','Toyota', 'Jeep','Hummer', 'Hyundai','Porsche']
plotje=together.plot()
plt.xticks(rotation=90)
plt.ylabel('Number of tweets')
plt.xlabel('Hour')
plt.savefig('carbrand_minute2.pdf', bbox_inches='tight')
```

```python
superbowl_c.columns = ['Super Bowl']
plotje=superbowl_c.plot()
plt.xticks(rotation=90)
plt.ylabel('Number of tweets')
plt.xlabel('Hour')
plt.savefig('superbowl_minute2.pdf', bbox_inches='tight')
```
