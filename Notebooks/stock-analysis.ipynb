```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
```

```python
import numpy as np
import csv
```

```python
from datetime import datetime
```

```python
fileNames = ['Audi.csv', 'Hummer.csv', 'Hyundai.csv', 'Jeep.csv', 'KIA.csv', 'Porsche.csv', 'Toyota.csv']
```

```python
priceWaves = [[],[],[],[],[],[],[]]
dates = [[],[],[],[],[],[],[]]
for i in range(0,len(fileNames)-1):
    print(i)
    with open('../Datasets/' + fileNames[i]) as f:
        title = fileNames[i].split('.')[0] + " Stock Price"
        imgName = fileNames[i].split('.')[0]+'StockPrice'
        reader = csv.reader(f)
        header_row = next(reader)
        startDate = datetime.strptime("2020-01-06", '%Y-%m-%d').date()
        midDate = datetime.strptime("2020-02-03", '%Y-%m-%d').date()
        gameDate = datetime.strptime("2020-02-02", '%Y-%m-%d').date()
        endDate =  datetime.strptime("2020-02-27", '%Y-%m-%d').date()
        date, high, low, priceWave = [], [], [], []
        for row in reader:
            current_date = datetime.strptime(row[0], '%Y-%m-%d').date()
            if (current_date >= startDate and current_date <= endDate):
                date.append(current_date)
                high.append(float(row[2]))
                low.append(float(row[3]))
                priceWave.append(float(row[4]) - float(row[1]))
                #priceWave.append(float(row[2]) - float(row[3]))

        fig = plt.figure(dpi=128, figsize=(10, 6))
        ax = plt.gca()
        plt.xticks(date, date[::1], fontsize=6, rotation=90)
        plt.axvline(startDate, alpha=0.1, c = 'black')
        plt.axvline(midDate, alpha=0.1, c = 'red')
        plt.axvline(gameDate, alpha=0.1, c = 'g')
        plt.axvline(endDate, alpha=0.1, c = 'black')
        plt.plot(date, high, c='red', alpha=0.5, label = 'High price', linewidth=1.5, marker='.')
        plt.plot(date, low, c='blue', alpha=0.5, label = 'Low price', linewidth=1.5, marker='.')
        plt.fill_between(date, high, low, facecolor='blue', alpha=0.1)
        plt.title(title, fontsize = 20)
        plt.xlabel('Date(2020-01-06~2020-02-28)', fontsize=15)
        fig.autofmt_xdate()
        plt.ylabel("Price($)", fontsize=15)
        ax = fig.add_subplot(111)
        ax.legend(loc='upper right')
        #plt.savefig(imgName + '.png', bbox_inches='tight')
        plt.show()
        priceWaves[i] = priceWave
        dates[i] = date
```
