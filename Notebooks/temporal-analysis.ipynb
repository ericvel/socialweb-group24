```python
#Date interval

audi = pd.read_csv("../Datasets/Audi.csv", error_bad_lines=False, sep=',',low_memory=False)
audi.drop('Unnamed: 0', axis=1, inplace=True)

kia = pd.read_csv("../Datasets/KIA.csv", error_bad_lines=False, sep=',', low_memory=False)
kia.drop('Unnamed: 0', axis=1, inplace=True)

toyota = pd.read_csv("../Datasets/Toyota.csv", error_bad_lines=False, sep=',',low_memory=False)
toyota.drop('Unnamed: 0', axis=1, inplace=True)

jeep = pd.read_csv("../Datasets/Jeep.csv", error_bad_lines=False, sep=',', low_memory=False)
jeep.drop('Unnamed: 0', axis=1, inplace=True)

hummer = pd.read_csv("../Datasets/Hummer.csv", error_bad_lines=False, sep=',', low_memory=False)
hummer.drop('Unnamed: 0', axis=1, inplace=True)

hyundai = pd.read_csv("../Datasets/Hyundai.csv", error_bad_lines=False, sep=',', low_memory=False)
hyundai.drop('Unnamed: 0', axis=1, inplace=True)

porsche = pd.read_csv("../Datasets/Porsche.csv", error_bad_lines=False, sep=',', low_memory=False)
porsche.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
start_date = '2020-01-01'
end_date = '2020-03-25'
```

```python
mask_k = (kia['date'] > start_date) & (kia['date'] <= end_date)
kia_i = kia.loc[mask_k]

mask_a = (audi['date'] > start_date) & (audi['date'] <= end_date)
audi_i = audi.loc[mask_a]

mask_t = (toyota['date'] > start_date) & (toyota['date'] <= end_date)
toyota_i = toyota.loc[mask_t]

mask_j = (jeep['date'] > start_date) & (jeep['date'] <= end_date)
jeep_i = jeep.loc[mask_j]

mask_hu = (hummer['date'] > start_date) & (hummer['date'] <= end_date)
hummer_i = hummer.loc[mask_hu]

mask_hy = (hyundai['date'] > start_date) & (hyundai['date'] <= end_date)
hyundai_i = hyundai.loc[mask_hy]

mask_p = (porsche['date'] > start_date) & (porsche['date'] <= end_date)
porsche_i = porsche.loc[mask_p]
```

```python
audi_i['date'] = pd.to_datetime(audi_i['date'], format='%Y-%m-%d', errors='coerce')
audi_i.set_index('date', inplace=True)
audi_i.index = audi_i.index.normalize()

kia_i['date'] = pd.to_datetime(kia_i['date'], format='%Y-%m-%d', errors='coerce')
kia_i.set_index('date', inplace=True)
kia_i.index = kia_i.index.normalize()

toyota_i['date'] = pd.to_datetime(toyota_i['date'], format='%Y-%m-%d', errors='coerce')
toyota_i.set_index('date', inplace=True)
toyota_i.index = toyota_i.index.normalize()

jeep_i['date'] = pd.to_datetime(jeep_i['date'], format='%Y-%m-%d', errors='coerce')
jeep_i.set_index('date', inplace=True)
jeep_i.index = jeep_i.index.normalize()

hummer_i['date'] = pd.to_datetime(hummer_i['date'], format='%Y-%m-%d', errors='coerce')
hummer_i.set_index('date', inplace=True)
hummer_i.index = hummer_i.index.normalize()

hyundai_i['date'] = pd.to_datetime(hyundai_i['date'], format='%Y-%m-%d', errors='coerce')
hyundai_i.set_index('date', inplace=True)
hyundai_i.index = hyundai_i.index.normalize()

porsche_i['date'] = pd.to_datetime(porsche_i['date'], format='%Y-%m-%d', errors='coerce')
porsche_i.set_index('date', inplace=True)
porsche_i.index = porsche_i.index.normalize()
```

```python
kia_int=pd.DataFrame(kia_i.groupby('date').count()['username'])
audi_int=pd.DataFrame(audi_i.groupby('date').count()['username'])
toyota_int=pd.DataFrame(toyota_i.groupby('date').count()['username'])
jeep_int=pd.DataFrame(jeep_i.groupby('date').count()['username'])
hummer_int=pd.DataFrame(hummer_i.groupby('date').count()['username'])
hyundai_int=pd.DataFrame(hyundai_i.groupby('date').count()['username'])
porsche_int=pd.DataFrame(porsche_i.groupby('date').count()['username'])
```

```python
together2 = pd.concat([kia_int, audi_int, toyota_int, jeep_int, hummer_int, hyundai_int, porsche_int], axis=1).fillna(0)
together2.columns = ['Kia','Audi','Toyota', 'Jeep','Hummer', 'Hyundai','Porsche']
plotje=together2.plot()
plt.savefig('carbrand2.pdf', bbox_inches='tight')
```
